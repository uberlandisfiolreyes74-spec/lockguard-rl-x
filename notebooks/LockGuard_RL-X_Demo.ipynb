## 1. Ejecutar Benchmark Simple

```python
print("🚀 Iniciando LockGuard RL-X Benchmark Demo")
print("="*60)

n_agents = 8
steps = 500

print(f"Agentes: {n_agents}")
print(f"Pasos por episodio: {steps}")
print("\nSimulando entorno financiero con gobernanza...")

# Simulación simple
risk_scores = np.random.uniform(0.1, 0.9, steps)
interventions = (risk_scores > 0.65).sum()

print(f"Eventos de alto riesgo detectados: {interventions}")
print(f"Tasa de intervención: {interventions/steps:.1%}")
print("\n✅ Demo completada")

plt.figure(figsize=(10,5))
plt.plot(risk_scores, label='AII (Adversarial Irreversibility Index)')
plt.axhline(y=0.65, color='r', linestyle='--', label='Umbral de intervención')
plt.title('Evolución del Riesgo de Irreversibilidad')
plt.xlabel('Pasos')
plt.ylabel('AII')
plt.legend()
plt.grid(True)
plt.show()
```
